In this notebook we quantify the amount of column reductions performed for producing the induced partial matching.

```python
import numpy as np
import tdqual.topological_data_quality_0 as tdqual

import matplotlib.pyplot as plt
import matplotlib as mpl

import os

plots_dir = "columns"
os.makedirs(plots_dir, exist_ok=True)
```

```python
RandGen = np.random.default_rng(2)
# # Generate Random Sample
Z = tdqual.sampled_circle(0,2,60, RandGen)
X_indices = RandGen.choice(Z.shape[0], 30, replace=False)
Z[X_indices]+=[0,0]
X = Z[X_indices]
# Plot point cloud
fig, ax = plt.subplots(ncols=1, figsize=(3,3))
ax.scatter(X[:,0], X[:,1], color=mpl.colormaps["RdBu"](0.3/1.3), s=60, marker="o", zorder=2)
ax.scatter(Z[:,0], Z[:,1], color=mpl.colormaps["RdBu"](1/1.3), s=40, marker="x", zorder=1)
ax.set_axis_off()
plt.savefig(plots_dir + "points_0.png")
```

```python
# Sort indices of points in Z, so that 
X_compl = np.ones(Z.shape[0], dtype="bool")
X_compl[X_indices] = False
Z = np.vstack((Z[X_indices], Z[X_compl]))
X_indices = range(len(X_indices))
X = Z[X_indices]
```

Podemos calcular el matching como antes

```python
from tdqual.topological_data_quality_0 import compute_Mf_0

filtration_list_X, filtration_list_Z, matching = compute_Mf_0(X, Z, is_dist=False)
```

```python
np.array(matching)
```

Tenemos una nueva opción que nos permite calcular el matching y obtener los índices de las columnas que se han sumado a cada una.

```python
from tdqual.topological_data_quality_0 import compute_Mf_0

filtration_list_X, filtration_list_Z, matching, reductions = compute_Mf_0(X, Z, is_dist=False, track_reductions=True)
```

```python
np.array(matching)
```
